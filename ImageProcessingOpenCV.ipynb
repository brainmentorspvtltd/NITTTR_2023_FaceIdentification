### Interpolation 

```python
interpolation_params = [cv2.INTER_AREA, cv2.INTER_CUBIC, cv2.INTER_LINEAR]
img_1 = cv2.resize(img, None, fx=1.0, fy=1.0, interpolation=interpolation_params[0])
img_2 = cv2.resize(img, None, fx=1.0, fy=1.0, interpolation=interpolation_params[1])
img_3 = cv2.resize(img, None, fx=1.0, fy=1.0, interpolation=interpolation_params[2])

images = [img, img_1, img_2, img_3]
titles = ['original', 'inter_area', 'inter_cubic', 'inter_linear']

for i in range(4):
#     cv2.imshow(titles[i], images[i])
#     cv2.waitKey(0)

    plt.subplot(2,2,i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

### Blurring 

```python
# Averaging blur
img = cv2.imread('img.webp')
blur_img_1 = cv2.blur(img, (5,5))
blur_img_2 = cv2.blur(img, (9,9))
blur_img_3 = cv2.blur(img, (11,11))
blur_img_4 = cv2.blur(img, (15,15))
blur_images = [blur_img_1, blur_img_2, blur_img_3, blur_img_4]
titles = ['5x5', '9x9', '11x11', '15x15']

for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(blur_images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
# Gaussian blur
img = cv2.imread('img.webp')
blur_img_1 = cv2.GaussianBlur(img, (5,5), 0)
blur_img_2 = cv2.GaussianBlur(img, (9,9), 0)
blur_img_3 = cv2.GaussianBlur(img, (11,11), 0)
blur_img_4 = cv2.GaussianBlur(img, (15,15), 0)
blur_images = [blur_img_1, blur_img_2, blur_img_3, blur_img_4]
titles = ['5x5', '9x9', '11x11', '15x15']

for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(blur_images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
# Median Blurring
blur_img_1 = cv2.medianBlur(img, 3)
blur_img_2 = cv2.medianBlur(img, 5)
blur_img_3 = cv2.medianBlur(img, 9)
blur_img_4 = cv2.medianBlur(img, 15)
blur_images = [blur_img_1, blur_img_2, blur_img_3, blur_img_4]
titles = ['3', '5', '9', '11']

for i in range(4):
    cv2.imshow('res', blur_images[i])
    cv2.waitKey(0)
#     plt.subplot(2,2,i+1)
#     plt.imshow(blur_images[i], 'gray')
#     plt.title(titles[i])
#     plt.xticks([])
#     plt.yticks([])
# plt.show()
```

```python
# Bilateral Filtering - used for removing noise without smoothing the edges
blur_img_1 = cv2.bilateralFilter(img, 11, 20, 10)
blur_img_2 = cv2.bilateralFilter(img, 11, 40, 20)
blur_img_3 = cv2.bilateralFilter(img, 11, 60, 30)
blur_img_4 = cv2.bilateralFilter(img, 11, 80, 40)
blur_images = [blur_img_1, blur_img_2, blur_img_3, blur_img_4]
titles = ['3', '5', '9', '11']

for i in range(4):
    cv2.imshow('res', blur_images[i])
    cv2.waitKey(0)
```

```python
# Thresholding
x, img_1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
x, img_2 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
x, img_3 = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
x, img_4 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)

thresh_images = [img_1, img_2, img_3, img_4]
titles = ['Binary', 'Binary Inv', 'Trunc', 'To Zero']

for i in range(4):
    cv2.imshow('res', thresh_images[i])
    cv2.waitKey(0)
```
